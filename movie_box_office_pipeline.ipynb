```python
URL = f"http://www.omdbapi.com/?apikey={OMDB_API_KEY}"
def fetch_movie_data(title):
    url_title = f"{URL}&t={title}"
    response = requests.get(url_title)
    data = response.json()
    
    if data.get("Response") == "False":
        print("Brak danych")

    return data
```

```python
unique_titles = df['title'].unique()
unique_titles_3 = unique_titles[0:3]
unique_titles_3
```

```python
movie_data = []

for title in unique_titles_3:
    data = fetch_movie_data(title)
    movie_data.append(data)    
```

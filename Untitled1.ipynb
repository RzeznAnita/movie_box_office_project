```python
from sqlalchemy import create_engine, String, Integer, Date, Float, Text, inspect, text

conn = create_engine("sqlite:///movies.db")
```

```python
inspector = inspect(engine)

tables = inspector.get_table_names()
print(tables)
```

```python
dim_movies = conn.execute(text("SELECT * FROM dim_movies"))
```
